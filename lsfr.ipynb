```python
#initalConfig integer < 2^m: listCs the feedback polynomial as a list in same order as in the book.
def lsfr(initialConfig, listCs):
    newConfig = initialConfig
    count = 0
    while True:
        outputBit = newConfig & 1
        xored = 0
        for i in listCs[0:len(listCs)-1]:
            xored = xored ^ ((newConfig >> i) & 1)
        #print (newConfig)
        print (format(newConfig,'0{}b'.format(listCs[-1])), outputBit)
        newConfig = (newConfig >> 1)|(xored << (listCs[-1]-1))
        count = count + 1 
        if (newConfig == initialConfig):
            break
    print('length of Seq:', count)
```

```python
lsfr(15, [0,1,4])
```

x^4 + x^1 + 1  : Primitive polynomial generating max-length sequence 15. 

```python
for i in range(1,16):
    lsfr(i, [0,2,4])
    print('\n')
    
```

x^4 + x^2 + 1: reducible polynomial, length of sequence depends on input 

(x^2 + x + 1)(x^2 -x + 1) Two different length sequences given above for initialization of 1111 and 0110.

```python
for i in range(1,16):
    lsfr(i , [0,1,2,3,4])
    print ('\n')
```
